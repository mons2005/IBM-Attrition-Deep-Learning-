```python
# prompt: explore attrition by age

sns.boxplot(x = 'Attrition', y = 'Age', data = IB)
```

Age Distribution by Attrition Status - The box plot illustrates the age distribution of employees who have left the company (Attrition = 'Yes') compared to those who have stayed (Attrition = 'No').

```python
# prompt: explore data for left employees

left = IB[IB['Attrition'] == 'Yes']
left.describe()
```

Descriptive Statistics for Employees Who Have Left the Company - This table provides summary statistics for employees with Attrition status set to 'Yes'.

```python
# prompt: find out  the distribution of emoloyees by  the education field

import matplotlib.pyplot as plt
sns.countplot(x = 'EducationField', data = left)
plt.xticks(rotation = 90)
plt.show()
```

 Distribution of Education Fields Among Employees Who Have Left the Company - The count plot displays the number of employees who have left IBM, categorized by their education field. The x-axis represents different education fields, and the y-axis shows the count of employees in each category.
